```python
from openai import OpenAI

# Inicializa cliente (aseg칰rate de tener tu API key en .env o en variable de entorno)
client = OpenAI()

def chat():
    # Mensaje de sistema: define el rol/persona del bot
    mensajes = [
        {
            "role": "system",
            "content": (
                "Eres un experto en tecnolog칤a que explica conceptos complejos "
                "de manera simple y amigable. Siempre incluyes ejemplos pr치cticos."
            )
        }
    ]

    print("Chatbot iniciado 游뱄 (escribe 'salir' para terminar)\n")

    while True:
        # Input del usuario
        user_input = input("T칰: ")
        if user_input.lower() == "salir":
            print("Chatbot finalizado 游녦")
            break

        # Se agrega a la conversaci칩n
        mensajes.append({"role": "user", "content": user_input})

        try:
            # Llamada a la API con todo el historial
            response = client.chat.completions.create(
                model="gpt-4o",
                messages=mensajes,
                temperature=0.7,
                max_tokens=300
            )

            # Respuesta del asistente
            reply = response.choices[0].message.content
            print("Bot:", reply, "\n")

            # Guardar respuesta en el historial
            mensajes.append({"role": "assistant", "content": reply})

        except Exception as e:
            print(f"Error: {e}")

# Ejecutar el chatbot
if __name__ == "__main__":
    chat()
```

## Explorando Par치metros de Configuraci칩n

Los par치metros m치s importantes al hacer llamadas a LLMs son:

- **temperature**: Controla la creatividad (0.0 = determin칤stico, 1.0 = muy creativo)
- **max_tokens**: L칤mite de tokens en la respuesta
- **model**: El modelo espec칤fico a usar (gpt-4o, gpt-3.5-turbo, etc.)
- **messages**: Array de mensajes con roles (system, user, assistant)

```python
# Comparando diferentes valores de temperature
def comparar_temperature():
    prompt = "Escribe una historia muy corta sobre un robot que aprende a cocinar."
    
    temperatures = [0.1, 0.5, 0.9]
    
    for temp in temperatures:
        print(f"\n{'='*50}")
        print(f"TEMPERATURE: {temp}")
        print('='*50)
        
        try:
            response = client.chat.completions.create(
                model="gpt-4o",
                messages=[{"role": "user", "content": prompt}],
                temperature=temp,
                max_tokens=100
            )
            
            print(response.choices[0].message.content)
            print(f"\nTokens usados: {response.usage.total_tokens}")
            
        except Exception as e:
            print(f"Error: {e}")

# Ejecutar comparaci칩n
comparar_temperature()
```
